```python
import os
import ast
import wfdb
import torch
import numpy as np
import pandas as pd
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tsai.all import InceptionTime
from captum.attr import LayerGradCam
import matplotlib.pyplot as plt

# ==========================================
# 1. Configuration & Paths
# ==========================================
DATA_PATH = '/content/kaggle/input/datasets/garethwmch/ptb-xl-1-0-3/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3'
SAMPLING_RATE = 100
BATCH_SIZE = 32
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ==========================================
# 2. Data Loading & Preprocessing
# ==========================================
def load_raw_data(df, sampling_rate, path):
    if sampling_rate == 100:
        data = [wfdb.rdsamp(os.path.join(path, f)) for f in df.filename_lr]
    else:
        data = [wfdb.rdsamp(os.path.join(path, f)) for f in df.filename_hr]
    data = np.array([signal for signal, meta in data])
    return data

# --- FIX: Download Kaggle dataset and ensure correct path usage ---
print("Downloading Kaggle dataset...")
# Ensure you have your kaggle.json uploaded to Colab for authentication
!mkdir -p ~/.kaggle
!cp /content/kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json
!kaggle datasets download -d garethwmch/ptb-xl-1-0-3
!mkdir -p /content/kaggle/input/datasets/garethwmch/ptb-xl-1-0-3/
!unzip -q ptb-xl-1-0-3.zip -d /content/kaggle/input/datasets/garethwmch/ptb-xl-1-0-3/
print("Kaggle dataset downloaded.")

print("Loading dataset metadata...")
# Use os.path.join for robustness
Y = pd.read_csv(os.path.join(DATA_PATH, 'ptbxl_database.csv'), index_col='ecg_id')
Y.scp_codes = Y.scp_codes.apply(lambda x: ast.literal_eval(x))

Y['age'] = Y['age'].fillna(Y['age'].median())
Y['sex'] = Y['sex'].fillna(0)
Y['height'] = Y['height'].fillna(Y['height'].median())
Y['weight'] = Y['weight'].fillna(Y['weight'].median())

def aggregate_diagnostic(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_class)
    return list(set(tmp))

agg_df = pd.read_csv(os.path.join(DATA_PATH, 'scp_statements.csv'), index_col=0)
agg_df = agg_df[agg_df.diagnostic == 1]
Y['diagnostic_superclass'] = Y.scp_codes.apply(aggregate_diagnostic)

Y['is_sick'] = Y['diagnostic_superclass'].apply(lambda x: 0 if 'NORM' in x and len(x) == 1 else 1)

print("Loading raw ECG signals...")
X = load_raw_data(Y, SAMPLING_RATE, DATA_PATH)

# ==========================================
# 3. Longitudinal Data Extraction (Trend Analysis)
# ==========================================
print("Extracting longitudinal data for trend analysis...")
Y['patient_id'] = Y['patient_id'].astype(int)

# Convert 'recording_date' column to datetime objects for proper sorting
Y['recording_date'] = pd.to_datetime(Y['recording_date'], errors='coerce')
Y = Y.sort_values(by=['patient_id', 'recording_date'])

# To get the previous ecg_id for longitudinal analysis,
# we need to shift the ecg_id (index) within each patient group.
# The previous approach `Y.groupby('patient_id').shift(1).index` was incorrect
# as it would return the original index.
Y['prev_ecg_id'] = Y.groupby('patient_id').apply(lambda x: x.index.to_series().shift(1)).droplevel(0)
valid_trend_indices = Y[Y['prev_ecg_id'].notna()].index

# ==========================================
# 4. Dataset Class (ECG + Clinical + Trend)
# ==========================================
class MultimodalECGDataset(Dataset):
    def __init__(self, ecg_data, clinical_data, labels, prev_ecg_data=None):
        self.ecg_data = torch.tensor(ecg_data, dtype=torch.float32).transpose(1, 2)
        self.clinical_data = torch.tensor(clinical_data, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.float32)

        if prev_ecg_data is not None:
            self.prev_ecg_data = torch.tensor(prev_ecg_data, dtype=torch.float32).transpose(1, 2)
        else:
            self.prev_ecg_data = torch.zeros_like(self.ecg_data)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.ecg_data[idx], self.prev_ecg_data[idx], self.clinical_data[idx], self.labels[idx]

clinical_features = ['age', 'sex', 'height', 'weight']
scaler = StandardScaler()
clinical_scaled = scaler.fit_transform(Y[clinical_features])

X_train_val, X_test, y_train_val, y_test, clin_train_val, clin_test = train_test_split(
    X, Y['is_sick'].values, clinical_scaled, test_size=0.2, random_state=42
)

train_dataset = MultimodalECGDataset(X_train_val, clin_train_val, y_train_val)
test_dataset = MultimodalECGDataset(X_test, clin_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# ==========================================
# 5. Fusion Model (Ready Architecture: InceptionTime)
# ==========================================
class ECGClinicalFusionModel(nn.Module):
    def __init__(self, clinical_dim=4):
        super(ECGClinicalFusionModel, self).__init__()

        # Using pre-built, highly reliable InceptionTime from tsai
        self.ecg_feature_extractor = InceptionTime(12, 1)
        self.ecg_feature_extractor.fc = nn.Identity()

        self.clinical_mlp = nn.Sequential(
            nn.Linear(clinical_dim, 16),
            nn.ReLU(),
            nn.Dropout(0.2)
        )

        self.trend_layer = nn.Sequential(
            nn.Linear(128 * 2, 128),
            nn.ReLU()
        )

        self.classifier = nn.Sequential(
            nn.Linear(128 + 16, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, curr_ecg, prev_ecg, clinical_data):
        curr_features = self.ecg_feature_extractor(curr_ecg)
        prev_features = self.ecg_feature_extractor(prev_ecg)

        trend_features = torch.cat((curr_features, prev_features), dim=1)
        ecg_combined = self.trend_layer(trend_features)

        clin_features = self.clinical_mlp(clinical_data)

        fused = torch.cat((ecg_combined, clin_features), dim=1)
        output = self.classifier(fused)
        return output

model = ECGClinicalFusionModel().to(DEVICE)

# ==========================================
# 6. Smart Diagnosis Logic & Uncertainty
# ==========================================
def get_diagnosis(probability, threshold_low=0.45, threshold_high=0.55):
    if probability < threshold_low:
        return "Healthy (Normal)"
    elif probability > threshold_high:
        return "Sick (Abnormal)"
    else:
        return "UNCERTAIN - Requires Human Doctor or immediate Echocardiogram"

# ==========================================
# 7. Training & Evaluation (Targeting > 95% Accuracy & F1)
# ==========================================
def evaluate_model(model, loader):
    model.eval()
    all_preds = []
    all_targets = []
    all_probs = []

    with torch.no_grad():
        for curr_ecg, prev_ecg, clin, target in loader:
            curr_ecg, prev_ecg, clin = curr_ecg.to(DEVICE), prev_ecg.to(DEVICE), clin.to(DEVICE)
            outputs = model(curr_ecg, prev_ecg, clin).squeeze()

            all_probs.extend(outputs.cpu().numpy())
            preds = (outputs > 0.5).float()
            all_preds.extend(preds.cpu().numpy())
            all_targets.extend(target.numpy())

    acc = accuracy_score(all_targets, all_preds)
    f1 = f1_score(all_targets, all_preds)
    return acc, f1, all_probs, all_targets

# NOTE: For Kaggle execution, insert your training loop here using torch.optim.Adam
# We mock the evaluation printout to demonstrate the metrics implementation.
print("Model initialized. Ready for training loop.")
# acc, f1, probs, targets = evaluate_model(model, test_loader)
# print(f"Test Accuracy: {acc*100:.2f}% | F1 Score: {f1:.4f}")

# ==========================================
# 8. Explainability: Grad-CAM for 1D ECG
# ==========================================
def generate_gradcam(model, curr_ecg, prev_ecg, clinical_data):
    model.eval()

    # Target the last convolutional layer in InceptionTime
    target_layer = model.ecg_feature_extractor.inceptionblock

    # Wrapper function for Captum
    def forward_func(x):
        return model(x, prev_ecg, clinical_data)

    layer_gc = LayerGradCam(forward_func, target_layer)

    curr_ecg.requires_grad_()
    attributions = layer_gc.attribute(curr_ecg, target=0)

    attributions = attributions.squeeze().cpu().detach().numpy()
    signal = curr_ecg.squeeze().cpu().detach().numpy()

    return signal, attributions

def plot_gradcam_heatmap(signal, attributions, lead=0):
    fig, ax = plt.subplots(figsize=(12, 4))

    time_axis = np.arange(signal.shape[1])
    ax.plot(time_axis, signal[lead], color='black', alpha=0.7, label='ECG Signal')

    attr_norm = np.maximum(attributions[lead], 0)
    attr_norm = attr_norm / (np.max(attr_norm) + 1e-8)

    im = ax.imshow(attr_norm[np.newaxis, :], cmap='jet', aspect='auto',
                   extent=[0, len(time_axis), np.min(signal[lead]), np.max(signal[lead])],
                   alpha=0.5)

    plt.colorbar(im, ax=ax, label='Importance (Grad-CAM)')
    plt.title(f"Grad-CAM Heatmap - Lead {lead}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import torch.optim as optim
import torch.nn as nn

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

EPOCHS = 10
print(f"Starting training on: {DEVICE}")

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0

    for curr_ecg, prev_ecg, clin, target in train_loader:
        curr_ecg = curr_ecg.to(DEVICE)
        prev_ecg = prev_ecg.to(DEVICE)
        clin = clin.to(DEVICE)
        target = target.to(DEVICE)

        optimizer.zero_grad()

        outputs = model(curr_ecg, prev_ecg, clin).squeeze()
        loss = criterion(outputs, target)

        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    acc, f1, _, _ = evaluate_model(model, test_loader)
    print(f"Epoch [{epoch+1}/{EPOCHS}] -> Loss: {train_loss/len(train_loader):.4f} | Acc: {acc*100:.2f}% | F1: {f1:.4f}")

print("Training Completed.")


sample_ecg, sample_prev, sample_clin, sample_label = next(iter(test_loader))
sample_ecg = sample_ecg[:1].to(DEVICE)
sample_prev = sample_prev[:1].to(DEVICE)
sample_clin = sample_clin[:1].to(DEVICE)

signal, attributions = generate_gradcam(model, sample_ecg, sample_prev, sample_clin)
plot_gradcam_heatmap(signal, attributions, lead=0)
```
